## Análise de Pinguins e História dos Oceanos Dataset do Kaggle: Palmer's Penguins
### Tarefas Realizadas
1. **Tratamento de Dados:**
   - Identificação e manipulação de valores ausentes ou inconsistentes.
2. **Análise Exploratória:**
   - Comparação de peso, comprimento do bico e comprimento da asa entre espécies.
   - Análise de como o habitat e a dieta influenciam as características físicas.
3. **Visualizações:**
   - Gráficos de dispersão, boxplots e histogramas para comparar as variáveis.
   - Narrativa baseada nos dados para interpretar as condições atuais dos pinguins.

```python
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go 
import seaborn as sns
import pandas as pd
```

```python

df_penguins = pd.read_csv('data_root/palmerpenguins_extended.csv')

# Verificar valores ausentes
valores_ausentes = df_penguins.isnull().sum()
print(valores_ausentes)

# Verificar inconsistências (como valores duplicados)
valores_duplicados = df_penguins.duplicated().sum()
print(f"Valores duplicados: {valores_duplicados}")
```

```python
# Agrupar por espécie e calcular as médias das características físicas
caracteristicas_fisicas = df_penguins.groupby('species')[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']].mean()
print(caracteristicas_fisicas)
```

```python
# Boxplot para mostrar a distribuição do comprimento do bico por espécie
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_penguins, x='species', y='bill_length_mm')
plt.title("Boxplot do Comprimento do Bico por Espécie")
plt.show()
```

```python
# Histograma para mostrar a distribuição do peso corporal por espécie
plt.figure(figsize=(10, 6))
sns.histplot(data=df_penguins, x='body_mass_g', hue='species', kde=True)
plt.title("Distribuição do Peso Corporal por Espécie")
plt.show()
```

```python
# Gráfico de barras empilhadas para comparar as médias das características físicas por espécie
caracteristicas_fisicas.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title("Comparação Empilhada das Características Físicas por Espécie")
plt.ylabel("Média das Características Físicas")
plt.show()
```

```python
# Boxen plot para comparar o peso corporal entre as espécies
plt.figure(figsize=(10, 6))
sns.boxenplot(data=df_penguins, x='species', y='body_mass_g')
plt.title("Boxen Plot do Peso Corporal por Espécie")
plt.show()
```

```python
# Violin plot para comparar o peso corporal por sexo e espécie
plt.figure(figsize=(10, 6))
sns.violinplot(data=df_penguins, x='species', y='body_mass_g', hue='sex', split=True)
plt.title("Peso Corporal por Sexo e Espécie")
plt.show()
```

```python
# Contar tipos de alimentos por espécie e ilha
diet_by_species_island = df_penguins.groupby(['species', 'island', 'diet']).size().unstack().fillna(0)

# Gráfico de barras empilhadas
diet_by_species_island.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab20')
plt.title("Distribuição da Dieta por Espécie e Ilha")
plt.xlabel("Espécie e Ilha")
plt.ylabel("Quantidade de Alimentos")
plt.legend(title="Tipo de Dieta")
plt.show()
```

#### Condições Físicas
As espécies de pinguins variam bastante em termos de características físicas, e isso pode estar relacionado ao ambiente de suas ilhas. Por exemplo, pinguins que vivem em ilhas mais frias, como a Biscoe, costumam ser maiores e com bicos mais longos, o que pode ajudar na caça em águas mais profundas. Já pinguins de ilhas com climas mais amenos parecem ser menos robustos, mostrando que as condições ambientais afetam diretamente suas características.

#### Dieta e Diferenças Entre Ilhas
A dieta dos pinguins também muda bastante entre as ilhas, com algumas espécies comendo mais crustáceos, peixes ou lulas, de acordo com o que está disponível. Isso deixa claro que o ambiente de cada ilha tem grande influência no que os pinguins comem e, consequentemente, em suas características físicas.
